```python
from IPython.display import display
from PIL import Image
from yolo import YOLO

def objectDetection(file, model_path, class_path, imgname):
    yolo = YOLO(model_path=model_path, classes_path=class_path, anchors_path="model_data/yolo_anchors.txt", image_name=imgname)
    image = Image.open(file)
    result_image = yolo.detect_image(image)
    display(result_image)
```

inference on face-detection weights

```python
# yolo_face_weights
```

```python
import random

with open("/home/shjeong/Data/HumanHeadDetection/HollywoodHeads/Splits/val.txt", 'r') as f:
    lines = f.readlines()

for i in random.sample(range(0,len(lines)),10):
    image_filepath = "/home/shjeong/Data/HumanHeadDetection/HollywoodHeads/JPEGImages/"
    image_filename = image_filepath + lines[i][:-1] + ".jpeg"
    objectDetection(image_filename, 'model_data/yolo_face_weights.h5', 'model_data/yolo_classes.txt',lines[i][:-1]+".jpeg")
```
